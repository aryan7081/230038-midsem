**Paper:** Multiple Incremental Decremental Learning of Support Vector Machines (Karasuyama & Takeuchi, NeurIPS 2009)

## Dataset Choice

We use **synthetic binary classification data** generated via `sklearn.datasets.make_classification`.

**What the dataset is:** A randomly generated dataset with configurable number of samples, features, and class separation. We use 300 samples, 4 features, 2 informative features, and 2 classes.

**Why it is a reasonable testbed:** The paper evaluates on (1) artificial 2D Gaussian mixture data (n=500) and (2) Fisher river time-series (n=1423). Our synthetic dataset belongs to the same problem type—binary classification with continuous features—suitable for SVM. It allows controlled experiments for incremental/decremental updates: we can split data into initial training set and points to add, then measure CPU time and breakpoints as in the paper. The method's core contribution (computational efficiency when adding/removing multiple points) does not depend on domain-specific structure.

**Limitations compared to the paper:** (1) Smaller than the paper's artificial data (300 vs 500); (2) No time-series structure like Fisher river; (3) Simpler feature structure. These limitations are acceptable for student-level reproduction focusing on the algorithmic contribution.

## Preprocessing Steps

- Generate data with fixed `random_state=42` for reproducibility.
- Convert labels from {0, 1} to {-1, +1} for SVM formulation (paper Section 2).
- No scaling applied (make_classification produces reasonably scaled features); the paper uses normalized features for Fisher river but not necessarily for artificial data.

```python
# Reproducibility: set random seed at the top of the notebook
RANDOM_STATE = 42
import numpy as np
np.random.seed(RANDOM_STATE)
```

```python
from sklearn.datasets import make_classification

# Generate dataset: 300 samples, 4 features, 2 informative, binary classification
X, y = make_classification(
    n_samples=300,
    n_features=4,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    random_state=RANDOM_STATE
)

# Convert labels from {0,1} to {-1,+1} for SVM (paper Section 2)
y = 2 * y - 1

print(f"Dataset shape: X={X.shape}, y={y.shape}")
print(f"Class distribution: {-1}: {(y==-1).sum()}, +1: {(y==1).sum()}")
```

```python
# Split into initial training set and points to add (for incremental learning)
n_initial = 200
n_add = 50

X_train = X[:n_initial]
y_train = y[:n_initial]
X_add = X[n_initial:n_initial + n_add]
y_add = y[n_initial:n_initial + n_add]

print(f"Initial training: {X_train.shape[0]} samples")
print(f"Points to add: {X_add.shape[0]} samples")
```
