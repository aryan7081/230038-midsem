# Task 1.2: Key Assumptions

**Paper:** Multiple Incremental Decremental Learning of Support Vector Machines (Karasuyama & Takeuchi, NeurIPS 2009)

---

## Assumption 1

**Assumption:** The kernel matrix K (and thus Q = y_i y_j K(x_i, x_j)) is positive definite. The paper explicitly states that if the kernel matrix is singular (e.g., when there are two or more identical data points in M), the algorithm may not work.

**Why the method needs it:** The linear system (7) requires inverting the matrix M, which contains the submatrix Q_M. If Q is singular, M may be singular and the update direction in Equation (10) cannot be computed. The paper uses a heuristic (adding a small constant to the diagonal) to circumvent this.

**Violation scenario:** A dataset with duplicate or near-duplicate training points (e.g., repeated measurements, or points that are identical up to numerical precision) would violate this assumption. The kernel matrix would have repeated rows/columns, leading to singularity.

**Paper reference:** Section 4, paragraph before experiments: "we assume that the kernel matrix K is positive definite. If the kernel matrix happens to be singular... our algorithm may not work."

---

## Assumption 2

**Assumption:** The update direction for adding points, Delta_alpha_A = eta(C*1 - alpha_A) (Equation 4), approximates the shortest path to the optimal solution. This holds exactly when all new points end at alpha_i = C at optimality.

**Why the method needs it:** The paper's efficiency claim (fewer breakpoints than coordinate-wise updates) relies on following a shorter path in the coefficient space. If the chosen direction deviates significantly from the true shortest path, the number of breakpoints may increase, reducing the computational advantage.

**Violation scenario:** When many of the newly added points end up as margin support vectors (0 < alpha_i < C) or outliers (alpha_i = 0) at optimality, the direction (4) is not the shortest path. The paper acknowledges this: "for some of alpha_A, the second assumption does not necessarily hold because we do not know the optimal alpha_A beforehand."

**Paper reference:** Section 3.4, paragraph discussing the second assumption: "if the point i in A which was located inside the margin before the update moved to M during the update... the path with respect to this parameter is not really the shortest one."

---

## Assumption 3

**Assumption:** The number of breakpoints is proportional to the total length of the path in the coefficient space. The paper uses this to argue that a shorter path implies fewer breakpoints and thus lower computational cost.

**Why the method needs it:** The main computational cost is solving the linear system (10) at each breakpoint (O(|M|^2) per breakpoint). If breakpoints were not roughly proportional to path length, the efficiency argument would not hold.

**Violation scenario:** In high-dimensional or complex kernel spaces, the critical regions (polygonal regions where M, I, O, A are constant) may have irregular shapes. Breakpoints could cluster densely in some regions and be sparse in others, breaking the proportionality.

**Paper reference:** Section 3.4, first assumption: "The number of breakpoints is proportional to the total length of the path." The paper also states "The first assumption means that the breakpoints are uniformly distributed on the path."
